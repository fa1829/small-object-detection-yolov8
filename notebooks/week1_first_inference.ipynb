```python
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt

# yolov8n = nano model — smallest and fastest, good for learning
model = YOLO("yolov8n.pt")
print("Model loaded. Classes it knows:", model.names)
```

## Option A — Use a URL image (easiest)

```python
# A busy street scene — good for seeing multiple detections
results = model("https://ultralytics.com/images/bus.jpg")
results[0].show()
```

## Option B — Use your own photo

```python
# Replace with any image path on your machine
results = model("IMG_1602.jpg")
results[0].show()
```

## Explore the output

```python
results = model("https://ultralytics.com/images/bus.jpg")
r = results[0]

print("Bounding boxes (xyxy):", r.boxes.xyxy)
print("Confidence scores:", r.boxes.conf)
print("Class IDs:", r.boxes.cls)

# Human-readable class names
for box, conf, cls in zip(r.boxes.xyxy, r.boxes.conf, r.boxes.cls):
    print(f"  {model.names[int(cls)]}: {conf:.2f} at {box.tolist()}")
```

```python
results = model("IMG_1602.jpg")
r = results[0]

print("Bounding boxes (xyxy):", r.boxes.xyxy)
print("Confidence scores:", r.boxes.conf)
print("Class IDs:", r.boxes.cls)

# Human-readable class names
for box, conf, cls in zip(r.boxes.xyxy, r.boxes.conf, r.boxes.cls):
    print(f"  {model.names[int(cls)]}: {conf:.2f} at {box.tolist()}")
```

## What to notice

- Each detection has a **bounding box** (x1, y1, x2, y2), a **confidence score** (0–1), and a **class label**
- The model was pretrained on COCO — 80 everyday object classes
- Try a crowded scene and notice which small objects it **misses** — that's the problem we solve in Week 4
